```python
def generate_shortdesc(text, max_words=15):
    if pd.isna(text) or text.strip() == "":
        return ""
    # split op punten om zinnen te vinden
    sentences = text.split('.')
    for s in sentences:
        s = s.strip()
        if len(s.split()) > 2:  # negeer hele korte zinnen
            # pak de eerste max_words woorden
            return ' '.join(s.split()[:max_words])
    # fallback: eerste max_words woorden van hele text
    return ' '.join(text.split()[:max_words])
```

```python
def generate_shortdesc(text, max_words=15):
    if pd.isna(text) or str(text).strip() == "":
        return ""
    sentences = text.split('.')
    for s in sentences:
        s = s.strip()
        if len(s.split()) > 2:  # negeer korte zinnen
            return ' '.join(s.split()[:max_words])
    return ' '.join(text.split()[:max_words])

# vervang alle 'ntb' of lege shortdescription
df['shortdescription'] = df.apply(
    lambda row: generate_shortdesc(row['description'])
                if str(row['shortdescription']).strip().lower() == 'ntb' or str(row['shortdescription']).strip() == ""
                else row['shortdescription'],
    axis=1
)
```
