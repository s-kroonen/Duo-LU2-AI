```python
import pandas as pd
import numpy as np

df = pd.read_csv("Uitgebreide_VKM_dataset.csv")

# Stap 1 — alles naar string voor consistente verwerking
df = df.astype(str)

# Stap 2 — lijst met verdachte / ongewenste waarden
weird_values = [
    "",           # lege strings
    "nan",        # NaN nadat alles string is
    "none",
    "null",
    "nvt",
    "volgt", 
    "[]",
    "ntb",
    "nader te bepalen",
    "nog niet bekend",
    "nadert te bepalen",
    "tbd"
]

# Lowercase-versie voor matching
df_lower = df.apply(lambda col: col.str.lower().str.strip())
df = df.apply(lambda col: col.str.lower().str.strip())

df = df.drop(columns=["Rood","Groen","Blauw","Geel"])

# Stap 3 — vervang ALLES wat in weird_values zit door "NTB"
for value in weird_values:
    if value == "[]":
        df[df_lower == value] = "['ntb']"
    else:
        df[df_lower == value] = "ntb"

# Stap 4 — extra: lege velden of whitespace → NTB
df = df.apply(lambda col: col.str.strip().replace("", "ntb"))

# Klaar — je dataset is nu volledig opgeschoond
df.head()

df.to_csv("Uitgebreide_VKM_dataset_zonder_weird_data.csv", index=False)
```

```python
import re

df = pd.read_csv("Uitgebreide_VKM_dataset_zonder_weird_data.csv")
def generate_short(description, content, max_len=250):
    """
    Maak een eenvoudige shortdescription in dezelfde stijl als de bestaande dataset.
    """
    text = f"{description.strip()} {content.strip()}".strip()

    # Verwijder dubbele spaties
    text = re.sub(r"\s+", " ", text)

    # Hou het kort (zoals echte shorts in dataset)
    if len(text) > max_len:
        text = text[:max_len].rsplit(" ", 1)[0] + "..."

    return text


# Vul shortdescription aan waar deze NTB is
mask = df["shortdescription"].str.lower().str.strip() == "ntb"

for idx in df[mask].index:
    desc = df.loc[idx, "description"]
    cont = df.loc[idx, "content"]
    df.loc[idx, "shortdescription"] = generate_short(desc, cont)

df[mask][["shortdescription", "description", "content"]].head(80)
```
